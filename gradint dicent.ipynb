```python
m = 9.01 
b = -4.74
lr =0.1

epochs = 20 

for i in range (epochs):
    loss_slope = -2 * np.sum( y - m * x.ravel() - b )
    b = b - (lr *  loss_slope)

    y_pred = m * x - b 

    plt.plot(x , y_pred)

plt.scatter(x,y)
```
