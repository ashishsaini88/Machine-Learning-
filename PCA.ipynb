```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv(r"C:\Users\KOIN\Downloads\titanic (1).csv")
df.head()
```

```python
numeric = df[["Age","Fare","Pclass","Parch","SibSp"]]

numeric = numeric.fillna(numeric.mean())

scaler = StandardScaler()
scaler_data = scaler.fit_transform(numeric)

```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaler_data)

explained = pca.explained_variance_ratio_

print("explained varince data is here :" ,explained)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0], pca_data[:,1] , c="red" , marker=0)

plt.title("PCA titenic dataset")
plt.xlabel("first principal component")
plt.ylabel("second principal component")
plt.show()
```

```python
print(f"the explained varience by first componenet is : {explained[0]:.2f}")
print(f"the explained varience by second componenet is : {explained[1]:.2f}")
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 
```

```python
iris = load_iris()

x= iris.data
y= iris.target
```

```python
pca = PCA(n_components=2)

pca_x = pca.fit_transform(x)
```

```python
plt.scatter(pca_x [:,0] , pca_x [:,1] , c=y , cmap= "viridis")

plt.title(" pca on iris dataset")
plt.xlabel("first principal component")
plt.ylabel("second principal component")

plt.show
```
